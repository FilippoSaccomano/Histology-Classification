# ðŸ”¬ AN2DL Challenge 2: Histopathology Image Classification

## ðŸ“‹ Project Overview
This notebook implements a sophisticated pipeline for **histopathology whole-slide image classification** using a hybrid TensorFlow-PyTorch approach:

- **Feature Extraction**: TensorFlow/Keras Path Foundation model (on CPU) to extract patch-level embeddings
- **Classification**: PyTorch Multi-Instance Learning (MIL) with Gated Attention mechanism
- **Architecture**: 4-class One-vs-Rest (OVR) classifier with ensemble training

## ðŸ—ï¸ Pipeline Architecture

### 1. **Patch Extraction & Filtering** 
   - Extract overlapping patches from large histopathology images
   - Intelligent tissue/background filtering using HSV color space
   - Black/ink region removal with adaptive thresholds
   - Top-k patch selection to manage computational resources

### 2. **Feature Embedding**
   - Path Foundation (TF-Keras) extracts 768-dim embeddings per patch
   - Embeddings cached as NPZ files for efficient reuse
   - Data augmentation (flip, rotate) applied during training

### 3. **Multi-Instance Learning**
   - Each image = "bag" of patch embeddings
   - Gated Attention aggregates patch-level information to image-level prediction
   - Squeeze-and-Excitation (SE) blocks for channel attention
   - R-Drop regularization for consistency

### 4. **Training & Evaluation**
   - Ensemble of 6 models with different random seeds
   - Class-weighted loss with label smoothing
   - Image-level stratified validation split
   - Macro-F1 score optimization

---

## âš™ï¸ Key Configuration
- **Patch size**: 224Ã—224 with stride 112 (50% overlap)
- **Max patches per image**: 12 (after filtering)
- **Embedding dimension**: 768 (Path Foundation output)
- **Hidden dimension**: 160 (MIL head)
- **Batch size**: 128 bags (images) per batch
- **Training epochs**: 70 (with early stopping)

## ðŸ” Section 5: Patch Extraction & Tissue Filtering

```python
import cv2

# ============ Patch extraction utilities ============

def pad_reflect_to_size(img_np: np.ndarray, target_h: int, target_w: int) -> np.ndarray:
    """Pad image to target size using reflection (for boundary patches)"""
    h, w, _ = img_np.shape
    pad_bottom = max(0, target_h - h)
    pad_right  = max(0, target_w - w)
    if pad_bottom == 0 and pad_right == 0:
        return img_np
    return cv2.copyMakeBorder(img_np, 0, pad_bottom, 0, pad_right, borderType=cv2.BORDER_REFLECT_101)

def compute_grid(H, W, patch_size, stride, drop_last_patch):
    """
    Compute grid of patch coordinates (x, y) for sliding window extraction
    
    Args:
        H, W: Image height and width
        patch_size: Size of square patches to extract
        stride: Step size between patches
        drop_last_patch: If True, drop patches that don't fit; if False, pad them
    
    Returns:
        xs, ys: Lists of x and y coordinates for top-left corners of patches
    """
    if drop_last_patch:
        xs = list(range(0, W - patch_size + 1, stride)) if W >= patch_size else []
        ys = list(range(0, H - patch_size + 1, stride)) if H >= patch_size else []
    else:
        n_cols = max(1, math.ceil((W - patch_size) / stride) + 1) if W > patch_size else 1
        n_rows = max(1, math.ceil((H - patch_size) / stride) + 1) if H > patch_size else 1
        xs = [min(k * stride, max(0, W - patch_size)) for k in range(n_cols)]
        ys = [min(k * stride, max(0, H - patch_size)) for k in range(n_rows)]
    return xs, ys

# ============ Tissue vs background filtering (HSV-based) ============
# These heuristics help identify tissue regions vs. background/ink in histopathology images

TF_INPUT_SIZE = 224
TISSUE_DOWNSAMPLE = 96  # Downsample patch for faster filtering

# HSV thresholds for tissue detection (more selective than default)
TISSUE_S_THR = 22   # Saturation threshold for tissue
TISSUE_V_MAX = 242  # Value (brightness) max for tissue

# Background detection (very white/desaturated regions)
BG_S_MAX = 10
BG_V_MIN = 246

# Ink/marker detection (very dark + saturated regions)
INK_V_MAX = 85
INK_S_MIN = 28

# Patch acceptance criteria
MIN_TISSUE_FRAC = 0.020  # Minimum tissue fraction to keep patch
MAX_INK_FRAC    = 0.220  # Maximum ink fraction allowed
MIN_SCORE_KEEP  = 0.010  # Minimum quality score to keep patch

def _resize_rgb(patch_rgb: np.ndarray, out_size: int = TF_INPUT_SIZE) -> np.ndarray:
    """Resize patch to target size for model input"""
    h, w = patch_rgb.shape[:2]
    if h == out_size and w == out_size:
        return patch_rgb
    interp = cv2.INTER_AREA if (h > out_size or w > out_size) else cv2.INTER_LINEAR
    return cv2.resize(patch_rgb, (out_size, out_size), interpolation=interp)

def _hsv_small(patch_rgb: np.ndarray):
    """Convert patch to HSV and downsample for efficient analysis"""
    hsv = cv2.cvtColor(patch_rgb, cv2.COLOR_RGB2HSV)
    if hsv.shape[0] != TISSUE_DOWNSAMPLE or hsv.shape[1] != TISSUE_DOWNSAMPLE:
        hsv = cv2.resize(hsv, (TISSUE_DOWNSAMPLE, TISSUE_DOWNSAMPLE), interpolation=cv2.INTER_AREA)
    return hsv

def _tissue_ink_bg_frac(patch_rgb_224: np.ndarray):
    """
    Compute fraction of pixels classified as tissue, ink, or background
    
    Returns:
        (tissue_frac, ink_frac, bg_frac): Fractions in [0, 1]
    """
    hsv = _hsv_small(patch_rgb_224)
    S = hsv[:, :, 1]
    V = hsv[:, :, 2]

    tissue = (S > TISSUE_S_THR) & (V < TISSUE_V_MAX)
    ink    = (V < INK_V_MAX) & (S > INK_S_MIN)
    bg     = (S < BG_S_MAX) & (V > BG_V_MIN)

    return float(tissue.mean()), float(ink.mean()), float(bg.mean())

def patch_black_score(patch_rgb_224: np.ndarray) -> float:
    """
    Compute quality score for patch (higher = more likely to contain useful tissue)
    
    Heuristic combines tissue presence, ink penalty, and background penalty
    """
    tissue_frac, ink_frac, bg_frac = _tissue_ink_bg_frac(patch_rgb_224)
    return float(tissue_frac - 0.70 * ink_frac + 0.03 * (1.0 - bg_frac))

def is_black_patch(patch_rgb_224: np.ndarray) -> bool:
    """
    Determine if patch should be rejected (too much background/ink, too little tissue)
    
    Returns:
        True if patch should be dropped, False if it should be kept
    """
    tissue_frac, ink_frac, bg_frac = _tissue_ink_bg_frac(patch_rgb_224)
    score = patch_black_score(patch_rgb_224)

    # Reject if too little tissue
    if tissue_frac < MIN_TISSUE_FRAC:
        return True
    # Reject if quality score too low
    if score < MIN_SCORE_KEEP:
        return True
    # Reject if too much ink with little tissue
    if (ink_frac > MAX_INK_FRAC) and (tissue_frac < 0.25):
        return True
    # Reject if almost entirely background
    if (bg_frac > 0.985) and (tissue_frac < 0.04):
        return True
    return False
```

```python
def build_patch_index(
    df_in,
    img_root,
    img_col,
    patch_size=224,
    stride=224,
    drop_last_patch=False,
    use_black_filter=True,
    topk=12,
    force_one=True,
    tf_input_size=224,
):
    """
    Build index of patch coordinates for all images in dataframe
    
    This function:
    1. Extracts all possible patches from each image
    2. Filters out background/ink patches using HSV heuristics
    3. Ranks patches by quality score
    4. Keeps top-k patches per image (for computational efficiency)
    5. Ensures at least one patch per image (if force_one=True)
    
    Args:
        df_in: DataFrame with image information
        img_root: Root directory containing images
        img_col: Column name with image filenames
        patch_size: Size of patches to extract
        stride: Stride between patches
        drop_last_patch: Whether to drop boundary patches
        use_black_filter: Whether to filter background/ink patches
        topk: Maximum patches to keep per image (None = keep all)
        force_one: Always keep at least one patch per image
        tf_input_size: Size to resize patches for filtering
    
    Returns:
        df_out: Filtered dataframe (only images with valid patches)
        coords: List of (img_id, x, y) tuples for all kept patches
    """
    df_in = df_in.reset_index(drop=True)
    valid_rows = []
    coords = []

    total_candidates = 0
    dropped_black = 0
    forced_imgs = 0

    for i in range(len(df_in)):
        img_name = str(df_in.iloc[i][img_col])
        img_path = os.path.join(img_root, img_name)
        if not os.path.isfile(img_path):
            continue

        try:
            img_np = np.array(Image.open(img_path).convert("RGB"))
        except Exception:
            continue

        H, W = img_np.shape[:2]
        xs, ys = compute_grid(H, W, patch_size, stride, drop_last_patch)

        cand = []
        all_cand = []

        # Extract and score all patches
        for y in ys:
            for x in xs:
                total_candidates += 1

                patch = img_np[y:y+patch_size, x:x+patch_size, :]
                if not drop_last_patch:
                    patch = pad_reflect_to_size(patch, patch_size, patch_size)

                patch_224 = _resize_rgb(patch, tf_input_size)
                score = patch_black_score(patch_224)
                all_cand.append((score, x, y))

                # Apply tissue/background filter
                if use_black_filter and is_black_patch(patch_224):
                    dropped_black += 1
                    continue

                cand.append((score, x, y))

        # Fallback: if all patches rejected, keep the best one
        if len(cand) == 0 and force_one and len(all_cand) > 0:
            all_cand.sort(key=lambda t: t[0], reverse=True)
            cand = [all_cand[0]]
            forced_imgs += 1

        # Skip images with no valid patches
        if len(cand) == 0:
            continue

        # Keep top-k patches (by quality score)
        cand.sort(key=lambda t: t[0], reverse=True)
        if topk is not None and len(cand) > topk:
            cand = cand[:topk]

        # Add to output
        valid_rows.append(df_in.iloc[i])
        img_id_new = len(valid_rows) - 1
        for _, x, y in cand:
            coords.append((img_id_new, x, y))

    df_out = pd.DataFrame(valid_rows).reset_index(drop=True)

    print(f"[build_patch_index] images: {len(df_out)} | patches kept: {len(coords)}")
    if total_candidates > 0:
        print(f"[build_patch_index] black dropped: {dropped_black}/{total_candidates} ({dropped_black/total_candidates:.1%})")
    if forced_imgs > 0:
        print(f"[build_patch_index] fallback forced 1 patch for {forced_imgs} images")

    return df_out, coords
```

```python
# Visualize which patches are kept vs dropped by the tissue filter
# This helps validate that our HSV-based filtering is working correctly
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

def _get_all_grid_patches_for_image(img_path, patch_size, stride, drop_last_patch):
    """Extract all patches from image and mark which would be dropped"""
    img_np = np.array(Image.open(img_path).convert("RGB"))
    H, W = img_np.shape[:2]
    xs, ys = compute_grid(H, W, patch_size, stride, drop_last_patch)

    items = []
    for y in ys:
        for x in xs:
            patch = img_np[y:y+patch_size, x:x+patch_size, :]
            if not drop_last_patch:
                patch = pad_reflect_to_size(patch, patch_size, patch_size)
            patch_224 = _resize_rgb(patch, TF_INPUT_SIZE)
            drop = is_black_patch(patch_224) if USE_BLACK_FILTER else False
            score = patch_black_score(patch_224)
            items.append((x, y, drop, score))
    return img_np, items

def _draw_rect(ax, x, y, ps, color, lw=2, alpha=0.9):
    """Draw rectangle on matplotlib axis"""
    ax.plot([x, x+ps, x+ps, x, x], [y, y, y+ps, y+ps, y],
            linewidth=lw, alpha=alpha, color=color)

def show_kept_dropped_random(df_in, img_root, img_col, n_show=3, seed=42, topk=TOPK_IF_TOO_MANY):
    """
    Visualize patch selection on random images
    Green boxes = kept patches (after top-k selection)
    Red boxes = dropped patches (background/ink filter)
    """
    rnd = random.Random(seed)
    idxs = list(range(len(df_in)))
    rnd.shuffle(idxs)
    idxs = idxs[:n_show]

    for k, i in enumerate(idxs, 1):
        img_name = str(df_in.iloc[i][img_col])
        img_path = os.path.join(img_root, img_name)
        if not os.path.isfile(img_path):
            print("Missing:", img_path)
            continue

        img_np, items = _get_all_grid_patches_for_image(
            img_path, PATCH_SIZE, PATCH_STRIDE, DROP_LAST_PATCH
        )

        kept = [(x,y,sc) for (x,y,drop,sc) in items if not drop]
        dropped = [(x,y,sc) for (x,y,drop,sc) in items if drop]

        kept_sorted = sorted(kept, key=lambda t: t[2], reverse=True)
        kept_topk = kept_sorted[:topk] if (topk is not None and len(kept_sorted) > topk) else kept_sorted

        # Print statistics
        print(f"\n=== Example {k}/{n_show} ===")
        print("File:", img_name)
        print(f"Grid patches: {len(items)} | kept (pre-topk): {len(kept)} | dropped: {len(dropped)} | kept (after topk): {len(kept_topk)}")
        if len(kept_sorted) > 0:
            print(f"Kept score range: {kept_sorted[-1][2]:.4f} .. {kept_sorted[0][2]:.4f}")

        plt.figure(figsize=(10, 8))
        plt.imshow(img_np)
        plt.title(f"Kept (GREEN) vs Dropped (RED) patches\n{img_name}")
        plt.axis("off")

        # Draw dropped patches first (red, thinner)
        for (x,y,sc) in dropped:
            _draw_rect(plt.gca(), x, y, PATCH_SIZE, color="red", lw=1, alpha=0.55)

        # Draw kept patches after top-k (green, thicker)
        for (x,y,sc) in kept_topk:
            _draw_rect(plt.gca(), x, y, PATCH_SIZE, color="lime", lw=2, alpha=0.90)

        plt.show()

# Show visualization on 3 random training images
show_kept_dropped_random(train_df, TRAIN_IMAGES_ROOT, IMG_COL, n_show=3, seed=SEED, topk=TOPK_IF_TOO_MANY)
```

## ðŸŽ¨ Section 7: Feature Extraction (Path Foundation Embeddings)

```python
from tqdm.auto import tqdm

def _patch_to_model_input(patch_rgb_224: np.ndarray) -> np.ndarray:
    """Normalize patch to [0, 1] for model input"""
    return patch_rgb_224.astype(np.float32) / 255.0

def _augment_patch_rgb(patch_rgb: np.ndarray, rng: np.random.Generator) -> np.ndarray:
    """
    Apply random augmentations to patch:
    - Random horizontal flip
    - Random vertical flip
    - Random 90Â° rotation (0Â°, 90Â°, 180Â°, 270Â°)
    """
    if rng.random() < 0.5:
        patch_rgb = np.flip(patch_rgb, axis=1)
    if rng.random() < 0.5:
        patch_rgb = np.flip(patch_rgb, axis=0)
    k = int(rng.integers(0, 4))
    if k:
        patch_rgb = np.rot90(patch_rgb, k, axes=(0, 1))
    return patch_rgb.copy()

@tf.function(reduce_retracing=True)
def tf_forward(batch_nhwc):
    """TensorFlow function for batched forward pass through Path Foundation model"""
    return path_model(batch_nhwc, training=False)

def extract_and_cache(
    df_in,
    coords,
    img_root,
    out_npz_path,
    has_labels: bool,
    apply_aug: bool = False,
    aug_prob: float = 0.7,
    seed: int = 42,
):
    """
    Extract Path Foundation embeddings for all patches and cache to NPZ file
    
    This function:
    1. Loads each image
    2. Extracts patches at specified coordinates
    3. Optionally applies augmentation (training only)
    4. Runs patches through Path Foundation model to get embeddings
    5. Saves embeddings, labels, and metadata to compressed NPZ file
    
    Args:
        df_in: DataFrame with image info
        coords: List of (img_id, x, y) patch coordinates
        img_root: Root directory for images
        out_npz_path: Output path for NPZ cache
        has_labels: Whether images have labels (train/val) or not (test)
        apply_aug: Apply augmentation (training only)
        aug_prob: Probability of augmentation per patch
        seed: Random seed
    """
    feats_list  = []
    labels_list = []
    imgid_list  = []

    filenames = df_in[IMG_COL].astype(str).tolist()

    # Group patches by image ID
    by_img = [[] for _ in range(len(df_in))]
    for (img_id, x, y) in coords:
        if 0 <= img_id < len(by_img):
            by_img[img_id].append((x, y))

    rng = np.random.default_rng(seed)

    # Process each image
    for img_id in tqdm(range(len(df_in)), desc=f"Extract -> {os.path.basename(out_npz_path)}"):
        img_name = filenames[img_id]
        img_path = os.path.join(img_root, img_name)

        try:
            img_np = np.array(Image.open(img_path).convert("RGB"))
        except Exception:
            continue

        # Extract patches for this image
        patches = []
        for (x, y) in by_img[img_id]:
            patch = img_np[y:y+PATCH_SIZE, x:x+PATCH_SIZE, :]
            if patch.size == 0:
                continue
            if not DROP_LAST_PATCH:
                patch = pad_reflect_to_size(patch, PATCH_SIZE, PATCH_SIZE)

            # Apply augmentation if training
            if apply_aug and (rng.random() < aug_prob):
                patch = _augment_patch_rgb(patch, rng)

            patch_224 = _resize_rgb(patch, TF_INPUT_SIZE)
            patches.append(_patch_to_model_input(patch_224))

        if len(patches) == 0:
            continue

        # Run patches through Path Foundation model in batches
        for s in range(0, len(patches), TF_BATCH):
            batch = np.stack(patches[s:s+TF_BATCH], axis=0).astype(np.float32, copy=False)
            emb = tf_forward(tf.convert_to_tensor(batch)).numpy()

            # Store embeddings as float16 to save memory
            feats_list.append(emb.astype(np.float16, copy=False))
            imgid_list.append(np.full((emb.shape[0],), img_id, dtype=np.int64))

            if has_labels:
                ylab = int(df_in.iloc[img_id][LABEL_COL_NUM])
                labels_list.append(np.full((emb.shape[0],), ylab, dtype=np.int64))

        cleanup_memory()

    # Concatenate all embeddings
    feats   = np.concatenate(feats_list, axis=0) if len(feats_list) else np.zeros((0, EMB_DIM), np.float16)
    img_ids = np.concatenate(imgid_list, axis=0) if len(imgid_list) else np.zeros((0,), np.int64)

    # Save to compressed NPZ file
    if has_labels:
        labels = np.concatenate(labels_list, axis=0) if len(labels_list) else np.zeros((0,), np.int64)
        np.savez_compressed(out_npz_path, feats=feats, labels=labels, img_ids=img_ids,
                            filenames=np.array(filenames, dtype=object))
    else:
        np.savez_compressed(out_npz_path, feats=feats, img_ids=img_ids,
                            filenames=np.array(filenames, dtype=object))

    print("Saved:", out_npz_path, "| feats:", feats.shape)
    return out_npz_path

# Build patch indices for train and validation sets
train_df2, train_coords = build_patch_index(
    train_df, TRAIN_IMAGES_ROOT, IMG_COL,
    patch_size=PATCH_SIZE, stride=PATCH_STRIDE, drop_last_patch=DROP_LAST_PATCH,
    use_black_filter=USE_BLACK_FILTER, topk=TOPK_IF_TOO_MANY, force_one=FORCE_AT_LEAST_ONE_PATCH
)
val_df2, val_coords = build_patch_index(
    val_df, TRAIN_IMAGES_ROOT, IMG_COL,
    patch_size=PATCH_SIZE, stride=PATCH_STRIDE, drop_last_patch=DROP_LAST_PATCH,
    use_black_filter=USE_BLACK_FILTER, topk=TOPK_IF_TOO_MANY, force_one=FORCE_AT_LEAST_ONE_PATCH
)

num_train_images = len(train_df2)
num_val_images   = len(val_df2)

# Extract and cache embeddings (or use cached if already exists)
if not os.path.isfile(TRAIN_EMB_PATH):
    extract_and_cache(train_df2, train_coords, TRAIN_IMAGES_ROOT, TRAIN_EMB_PATH,
                      has_labels=True, apply_aug=True, aug_prob=0.7, seed=SEED)
else:
    print("Using cached:", TRAIN_EMB_PATH)

if not os.path.isfile(VAL_EMB_PATH):
    extract_and_cache(val_df2, val_coords, TRAIN_IMAGES_ROOT, VAL_EMB_PATH,
                      has_labels=True, apply_aug=False, seed=SEED)
else:
    print("Using cached:", VAL_EMB_PATH)
```

```python
# ============================================================
# CELL 9 - LOAD EMBEDDINGS -> BAG (MIL) DATASET/LOADERS
#   - ogni sample = 1 immagine ("bag") con N patch embeddings
#   - train: random patch dropout + noise su embedding (anti-overfit)
# ============================================================
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

train_npz = np.load(TRAIN_EMB_PATH, allow_pickle=True)
val_npz   = np.load(VAL_EMB_PATH, allow_pickle=True)

Xtr = train_npz["feats"].astype(np.float32)
ytr_patch = train_npz["labels"].astype(np.int64)   # patch label (uguale per immagine nel file)
idtr = train_npz["img_ids"].astype(np.int64)

Xva = val_npz["feats"].astype(np.float32)
yva_patch = val_npz["labels"].astype(np.int64)
idva = val_npz["img_ids"].astype(np.int64)

print("Train patches:", Xtr.shape, "| Val patches:", Xva.shape)

# ---- build contiguous segments per image (sorting by img_id) ----
def build_segments(X, y_patch, img_ids):
    order = np.argsort(img_ids)
    Xs = X[order]
    ys = y_patch[order]
    ids = img_ids[order]

    uniq, start_idx = np.unique(ids, return_index=True)
    start_idx = start_idx.astype(np.int64)
    end_idx = np.r_[start_idx[1:], np.array([len(ids)], dtype=np.int64)]

    # image label = majority (qui Ã¨ quasi sempre costante)
    y_img = []
    for s, e in zip(start_idx, end_idx):
        vals, cnt = np.unique(ys[s:e], return_counts=True)
        y_img.append(int(vals[np.argmax(cnt)]))
    y_img = np.array(y_img, dtype=np.int64)

    return Xs, ys, ids, uniq.astype(np.int64), start_idx, end_idx, y_img

Xtr_s, ytr_s, idtr_s, img_tr, st_tr, en_tr, yimg_tr = build_segments(Xtr, ytr_patch, idtr)
Xva_s, yva_s, idva_s, img_va, st_va, en_va, yimg_va = build_segments(Xva, yva_patch, idva)

num_train_images = len(img_tr)
num_val_images   = len(img_va)
print("Train images:", num_train_images, "| Val images:", num_val_images)

# ---- MIL bag dataset ----
KEEP_PATCH_FRAC_TRAIN = 0.5   # patch dropout dentro il bag (anti-overfit)
MIN_PATCHES_KEEP      = 3
EMB_NOISE_STD         = 0.05   # rumore su embedding (anti-overfit)

class BagEmbDataset(Dataset):
    def __init__(self, Xs, st, en, y_img=None, train_mode=False, seed=42):
        self.Xs = Xs
        self.st = st
        self.en = en
        self.y_img = y_img
        self.train_mode = train_mode
        self.rng = np.random.default_rng(seed)

    def __len__(self):
        return len(self.st)

    def __getitem__(self, idx):
        s = int(self.st[idx]); e = int(self.en[idx])
        bag = self.Xs[s:e]  # (P,D)

        if self.train_mode:
            P = bag.shape[0]
            if P > MIN_PATCHES_KEEP:
                k = max(MIN_PATCHES_KEEP, int(np.ceil(P * KEEP_PATCH_FRAC_TRAIN)))
                sel = self.rng.choice(P, size=k, replace=False)
                bag = bag[sel]

            if EMB_NOISE_STD > 0:
                noise = self.rng.normal(0.0, EMB_NOISE_STD, size=bag.shape).astype(np.float32)
                bag = bag + noise

        bag_t = torch.from_numpy(bag).float()
        if self.y_img is None:
            return bag_t, idx  # idx = image index
        return bag_t, int(self.y_img[idx]), idx

def collate_bags(batch):
    # batch: list of (bag, y, idx) oppure (bag, idx)
    has_label = (len(batch[0]) == 3)

    bags = [b[0] for b in batch]
    lengths = torch.tensor([x.shape[0] for x in bags], dtype=torch.long)
    B = len(bags)
    Pmax = int(lengths.max().item())
    D = int(bags[0].shape[1])

    X = torch.zeros((B, Pmax, D), dtype=torch.float32)
    mask = torch.zeros((B, Pmax), dtype=torch.bool)
    for i, bag in enumerate(bags):
        p = bag.shape[0]
        X[i, :p] = bag
        mask[i, :p] = True

    if has_label:
        y = torch.tensor([b[1] for b in batch], dtype=torch.long)
        idx = torch.tensor([b[2] for b in batch], dtype=torch.long)
        return X, mask, y, idx
    else:
        idx = torch.tensor([b[1] for b in batch], dtype=torch.long)
        return X, mask, idx

train_bag_ds = BagEmbDataset(Xtr_s, st_tr, en_tr, y_img=yimg_tr, train_mode=True, seed=SEED)
val_bag_ds   = BagEmbDataset(Xva_s, st_va, en_va, y_img=yimg_va, train_mode=False, seed=SEED)

# ---- image-level sampler (bilancia le immagini, non le patch) ----
USE_IMG_WEIGHTED_SAMPLER = False
BATCH_SIZE_BAG = 128 if n_gpu >= 2 else 64

if USE_IMG_WEIGHTED_SAMPLER:
    counts_img = np.bincount(yimg_tr, minlength=NUM_CLASSES).astype(np.float64)
    w_cls = counts_img.sum() / np.maximum(1.0, counts_img)
    w_samp = w_cls[yimg_tr]
    sampler = WeightedRandomSampler(torch.from_numpy(w_samp).double(),
                                    num_samples=len(w_samp),
                                    replacement=True)
    train_loader = DataLoader(train_bag_ds, batch_size=BATCH_SIZE_BAG, sampler=sampler,
                              num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY, collate_fn=collate_bags)
else:
    train_loader = DataLoader(train_bag_ds, batch_size=BATCH_SIZE_BAG, shuffle=True,
                              num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY, collate_fn=collate_bags)

val_loader = DataLoader(val_bag_ds, batch_size=BATCH_SIZE_BAG, shuffle=False,
                        num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY, collate_fn=collate_bags)

print("train bag batches:", len(train_loader), "| val bag batches:", len(val_loader))
```

## ðŸ§  Section 9: Gated Attention MIL Model Architecture

### 8.1 Bag Length Distribution

```python
"""
Multi-Instance Learning Model with Gated Attention

Architecture:
1. Patch Projection: Maps 768-dim embeddings to lower-dim hidden space
2. SE Block: Squeeze-and-Excitation for channel-wise attention
3. Gated Attention: Computes attention weights over patches (which patches are important)
4. Bag Aggregation: Weighted sum of patch features using attention weights
5. Classifier: Final linear layer for class prediction

Key features:
- Gated attention mechanism learns which patches matter most
- SE blocks add channel-wise recalibration
- Safe for mixed-precision training (critical operations in float32)
"""
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.amp import GradScaler

from lion_pytorch import Lion

class SEBlock(nn.Module):
    """
    Squeeze-and-Excitation Block for channel attention
    Recalibrates channel-wise feature responses
    """
    def __init__(self, channels, reduction=8):
        super().__init__()
        self.squeeze = nn.AdaptiveAvgPool1d(1)
        self.excitation = nn.Sequential(
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels, bias=False),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        """
        Args:
            x: [B, P, C] - Batch, Patches, Channels
        Returns:
            [B, P, C] - Recalibrated features
        """
        B, P, C = x.shape
        # Global average pooling across patches: [B, P, C] -> [B, C, P] -> [B, C, 1]
        y = x.transpose(1, 2)  # [B, C, P]
        y = self.squeeze(y).squeeze(-1)  # [B, C]
        
        # Excitation: compute channel importance
        y = self.excitation(y)  # [B, C]
        
        # Scale input by channel importance
        return x * y.unsqueeze(1)

class GatedAttention(nn.Module):
    """
    Gated Attention Mechanism for MIL
    
    Computes attention weights using gated combination:
    attention = w^T * (tanh(V*h) âŠ™ sigmoid(U*h))
    
    The gating mechanism allows the model to learn which features
    are relevant for attention computation
    """
    def __init__(self, in_dim, attn_dim):
        super().__init__()
        self.V = nn.Linear(in_dim, attn_dim)  # Value projection
        self.U = nn.Linear(in_dim, attn_dim)  # Gate projection
        self.w = nn.Linear(attn_dim, 1)       # Attention score

    def forward(self, H, mask):
        """
        Args:
            H:    [B, P, D] - Patch features
            mask: [B, P] bool - True for valid patches, False for padding
        Returns:
            alpha: [B, P] - Attention weights (sum to 1 per bag)
        """
        # IMPORTANT: Use float32 for attention to avoid numerical issues
        H32 = H.float()

        V = torch.tanh(self.V(H32))      # Value transform
        U = torch.sigmoid(self.U(H32))   # Gate transform
        A = self.w(V * U).squeeze(-1)    # Attention scores [B, P]

        # Mask out padding patches (set to large negative value)
        A = A.masked_fill(~mask, -1e4)

        # Softmax to get attention weights
        alpha = torch.softmax(A, dim=1)  # [B, P]
        return alpha

class GatedMIL(nn.Module):
    """
    Complete Gated MIL model for histopathology classification
    
    Args:
        emb_dim: Input embedding dimension (from Path Foundation)
        num_classes: Number of output classes
        hid: Hidden dimension for patch projection
        attn_dim: Attention mechanism dimension
        drop_patch: Dropout rate after patch projection
        drop_bag: Dropout rate after bag aggregation
        use_se: Whether to use SE blocks
        se_reduction: Reduction ratio for SE blocks
    """
    def __init__(self, emb_dim, num_classes, hid=192, attn_dim=128, 
                 drop_patch=0.25, drop_bag=0.35, use_se=True, se_reduction=8):
        super().__init__()
        
        # Patch-level feature transformation
        self.patch_proj = nn.Sequential(
            nn.LayerNorm(emb_dim),
            nn.Linear(emb_dim, hid),
            nn.GELU(),
            nn.Dropout(drop_patch),
            nn.LayerNorm(hid),
        )
        
        # SE block for channel attention (optional)
        self.use_se = use_se
        if self.use_se:
            self.se_block = SEBlock(hid, reduction=se_reduction)
        
        # Gated attention mechanism
        self.attn = GatedAttention(hid, attn_dim)
        
        # Bag-level aggregation and classification
        self.bag_drop = nn.Dropout(drop_bag)
        self.cls = nn.Linear(hid, num_classes)

    def forward(self, X, mask):
        """
        Args:
            X: [B, P, EMB_DIM] - Patch embeddings (may be fp16 under autocast)
            mask: [B, P] bool - Valid patch mask
        
        Returns:
            logits: [B, NUM_CLASSES] - Class logits
            alpha: [B, P] - Attention weights
        """
        # Project patches to hidden space
        H = self.patch_proj(X)  # [B, P, hid]
        
        # Apply SE channel attention if enabled
        if self.use_se:
            H = self.se_block(H)  # [B, P, hid]
        
        # Compute attention weights over patches
        alpha = self.attn(H, mask)  # [B, P] in float32
        
        # Aggregate patches using attention (weighted sum)
        M = torch.sum(H.float() * alpha.unsqueeze(-1), dim=1)  # [B, hid]
        
        # Apply dropout and classify
        M = self.bag_drop(M)
        logits = self.cls(M)  # [B, NUM_CLASSES]
        
        return logits, alpha

def build_model():
    """Build and initialize the Gated MIL model"""
    m = GatedMIL(
        EMB_DIM, NUM_CLASSES, 
        hid=160, attn_dim=96, 
        drop_patch=0.4, drop_bag=0.5, 
        use_se=True, se_reduction=8
    ).to(device)
    
    # Use DataParallel for multi-GPU training
    if n_gpu >= 2:
        m = nn.DataParallel(m)
        print(f"âœ“ DataParallel ({n_gpu} GPUs)")
    return m

def unwrap(m):
    """Unwrap model from DataParallel wrapper if needed"""
    return m.module if hasattr(m, "module") else m

# Initialize gradient scaler for mixed-precision training
scaler = GradScaler("cuda", enabled=torch.cuda.is_available())
```

```python
# Training configuration and loss function setup
import numpy as np
import torch
import torch.nn.functional as F
from sklearn.metrics import f1_score

# Compute class weights to handle class imbalance
counts = np.bincount(yimg_tr, minlength=NUM_CLASSES).astype(np.float64)
w = counts.sum() / np.maximum(1.0, counts)  # Inverse frequency weighting
w = (w / w.mean()).astype(np.float32)       # Normalize
print("image counts:", counts)
print("class weights:", w)

class_w = torch.tensor(w, dtype=torch.float32, device=device)

# Loss function with class weighting and label smoothing
LABEL_SMOOTH = 0.1  # Prevents overconfident predictions
criterion = torch.nn.CrossEntropyLoss(weight=class_w, label_smoothing=LABEL_SMOOTH)

# R-Drop regularization: consistency between two forward passes with different dropout
USE_RDROP = True
RDROP_ALPHA = 0.8  # Weight for KL divergence consistency loss

# Training hyperparameters
EPOCHS = 70
LR = 3e-4                  # Base learning rate
WEIGHT_DECAY = 2e-2        # L2 regularization
GRAD_CLIP = 1.0            # Gradient clipping for stability
PATIENCE = 18              # Early stopping patience (epochs without improvement)
MIN_DELTA = 0.0005         # Minimum improvement to reset patience

def build_optimizer(model):
    """Build Lion optimizer (memory-efficient, often outperforms AdamW)"""
    return Lion(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)

def build_scheduler(optimizer):
    """
    Build OneCycleLR scheduler
    - Warmup for 10% of training
    - Cosine annealing to final LR
    - Max LR is 6x base LR
    """
    total_steps = EPOCHS * len(train_loader)
    return torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=LR * 6.0,
        total_steps=total_steps,
        pct_start=0.10,          # 10% warmup
        anneal_strategy="cos",
        div_factor=25.0,
        final_div_factor=2000.0,
    )

@torch.no_grad()
def eval_val_f1(model, loader):
    """
    Evaluate model on validation set and compute macro F1 score
    
    Args:
        model: Trained model
        loader: Validation dataloader
    
    Returns:
        macro_f1: Macro-averaged F1 score across all classes
    """
    model.eval()
    ys, ps = [], []
    
    for X, mask, y, _ in loader:
        X = X.to(device, non_blocking=True)
        mask = mask.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        logits, _ = model(X, mask)
        pred = torch.argmax(logits, dim=1)

        ys.append(y.detach().cpu().numpy())
        ps.append(pred.detach().cpu().numpy())

    y_true = np.concatenate(ys)
    y_pred = np.concatenate(ps)
    return float(f1_score(y_true, y_pred, average="macro"))
```

## ðŸ‹ï¸ Section 11: Ensemble Training Loop

```python
# ============================================================
# CELL 12 - TRAIN ENSEMBLE (MIL, image-level) + PRINT TRAIN_F1
# ============================================================
import os
import numpy as np
import torch
from torch.amp import autocast
import torch.nn.functional as F
from sklearn.metrics import f1_score

SEEDS = [42, 1337, 2024, 22, 37, 78]
ensemble_ckpts = []
seed_scores = {}
seed_histories = {}

def set_seed(seed: int):
    import random
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

def save_ckpt(path: str, model):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    torch.save(unwrap(model).state_dict(), path)

def load_ckpt_into(model, path: str):
    sd = torch.load(path, map_location=device)
    sd = {k.replace("module.", ""): v for k, v in sd.items()}
    unwrap(model).load_state_dict(sd, strict=True)
    model.eval()
    return model

@torch.no_grad()
def eval_loader_f1(model, loader):
    model.eval()
    ys, ps = [], []
    for X, mask, y, _ in loader:
        X = X.to(device, non_blocking=True)
        mask = mask.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        with autocast("cuda", enabled=torch.cuda.is_available()):
            logits, _ = model(X, mask)
        pred = torch.argmax(logits, dim=1)

        ys.append(y.detach().cpu().numpy())
        ps.append(pred.detach().cpu().numpy())

    y_true = np.concatenate(ys)
    y_pred = np.concatenate(ps)
    return float(f1_score(y_true, y_pred, average="macro"))

for seed in SEEDS:
    print(f"\n==============================")
    print(f"SEED {seed}")
    print(f"==============================")
    set_seed(seed)
    cleanup_memory()

    model = build_model()
    optimizer = build_optimizer(model)
    scheduler = build_scheduler(optimizer)

    best_val = -1.0
    best_ep  = -1
    no_improve = 0
    history = {"train_loss": [], "train_f1": [], "val_f1": [], "lr": []}

    ckpt_path = os.path.join(CKPT_DIR, f"mil_gated_best_seed{seed}.pth")

    for epoch in range(EPOCHS):
        model.train()
        run_loss = 0.0
        n_seen = 0

        for X, mask, y, _ in train_loader:
            X = X.to(device, non_blocking=True)
            mask = mask.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            optimizer.zero_grad(set_to_none=True)

            if USE_RDROP:
                with autocast("cuda", enabled=torch.cuda.is_available()):
                    logits1, _ = model(X, mask)
                    logits2, _ = model(X, mask)

                    ce = 0.5 * (criterion(logits1, y) + criterion(logits2, y))

                    p1 = F.softmax(logits1, dim=1)
                    p2 = F.softmax(logits2, dim=1)
                    kl = 0.5 * (F.kl_div(p1.log(), p2, reduction="batchmean") +
                                F.kl_div(p2.log(), p1, reduction="batchmean"))

                    loss = ce + RDROP_ALPHA * kl
            else:
                with autocast("cuda", enabled=torch.cuda.is_available()):
                    logits, _ = model(X, mask)
                    loss = criterion(logits, y)

            scaler.scale(loss).backward()
            if GRAD_CLIP and GRAD_CLIP > 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)

            scaler.step(optimizer)
            scaler.update()
            scheduler.step()

            run_loss += float(loss.item()) * y.size(0)
            n_seen += y.size(0)

        train_loss = run_loss / max(1, n_seen)

        # IMPORTANT: calcolo train_f1 in eval mode (meno rumore)
        train_f1 = eval_loader_f1(model, train_loader)
        val_f1   = eval_loader_f1(model, val_loader)

        history["train_loss"].append(train_loss)
        history["train_f1"].append(train_f1)
        history["val_f1"].append(val_f1)
        history["lr"].append(float(optimizer.param_groups[0]["lr"]))

        print(f"Epoch {epoch+1:02d} | loss={train_loss:.4f} | tr_f1={train_f1:.4f} | val_f1={val_f1:.4f}")

        if val_f1 > best_val + MIN_DELTA:
            best_val = val_f1
            best_ep = epoch
            no_improve = 0
            save_ckpt(ckpt_path, model)
        else:
            no_improve += 1
            if no_improve >= PATIENCE:
                print("EARLY STOP")
                break

        cleanup_memory()

    seed_histories[seed] = history
    seed_scores[seed] = best_val
    ensemble_ckpts.append(ckpt_path)
    print(f"Best seed {seed}: epoch={best_ep+1} | best_val_f1={best_val:.4f} | ckpt={ckpt_path}")

print("\nENSEMBLE CKPTS:")
for p in ensemble_ckpts:
    print(" -", p)
```

```python
"""
Final Test Inference & Submission Generation

This cell:
1. Loads all trained ensemble models
2. Runs inference on test set
3. Averages predictions across ensemble members
4. Generates submission CSV file
"""
import numpy as np
import pandas as pd

@torch.no_grad()
def predict_test_probs_ensemble(ckpt_paths, loader, num_images):
    """
    Run ensemble inference on test set
    
    Args:
        ckpt_paths: List of checkpoint paths for ensemble members
        loader: Test dataloader
        num_images: Number of test images
    
    Returns:
        probs: [num_images, NUM_CLASSES] - Class probabilities (averaged over ensemble)
    """
    # Load all ensemble models
    models = []
    for p in ckpt_paths:
        m = build_model()
        m = load_ckpt_into(m, p)
        m.eval()
        models.append(m)

    # Initialize probability matrix
    probs = np.zeros((num_images, NUM_CLASSES), dtype=np.float64)

    # Run inference
    for X, mask, idx in loader:
        X = X.to(device, non_blocking=True)
        mask = mask.to(device, non_blocking=True)
        idx_np = idx.numpy().astype(np.int64)

        # Accumulate logits from all models
        lg_acc = None
        for m in models:
            logits, _ = m(X, mask)
            lg_acc = logits if lg_acc is None else (lg_acc + logits)
        
        # Average logits and convert to probabilities
        logits = lg_acc / len(models)
        p = torch.softmax(logits, dim=1).detach().cpu().numpy()
        probs[idx_np] = p

    return probs

# Run ensemble inference on test set
probs_img = predict_test_probs_ensemble(ensemble_ckpts, test_loader, num_test_images)

# Convert probabilities to class predictions
pred_id = np.argmax(probs_img, axis=1).astype(int)
pred_lbl = [id2label[i] for i in pred_id]

# Create submission dataframe
# IMPORTANT: Use test_df2 (after patch filtering) to ensure alignment
sub_df = pd.DataFrame({
    IMG_COL: test_df2[IMG_COL].astype(str).tolist(),
    LABEL_COL: pred_lbl
}).sort_values(IMG_COL).reset_index(drop=True)

# Save submission CSV
sub_df.to_csv(SUBMISSION_PATH, index=False)
print("Saved submission:", SUBMISSION_PATH)
print(sub_df[LABEL_COL].value_counts())
display(sub_df.head())
```
